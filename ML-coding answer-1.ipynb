```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd

# Dataset
data = {
    'Customer_ID': [101, 102, 103, 101, 104, 102, 101, 105, 102, 103],
    'Purchase_Amount': [200, 150, 180, 220, 300, 200, 100, 400, 250, 300]
}

# Create DataFrame
df = pd.DataFrame(data)

# Group by Customer_ID and sum Purchase_Amount
customer_totals = df.groupby('Customer_ID')['Purchase_Amount'].sum()

# Sort and get top 3 customers
top_customers = customer_totals.sort_values(ascending=False).head(3)

print("Total purchase per customer:")
print(customer_totals)
print("\nTop 3 customers with highest purchase amounts:")
print(top_customers)
```

```python

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


#Dataset:
data = {'Square_Feet': [1500, 2000, 2500, 3000, 3500],
        'Price': [300000, 400000, 500000, 600000, 700000]}

df = pd.DataFrame(data)
df

#splitting features & target
X = df[['Square_Feet']]
Y = df['Price']

#Training linear regression 
Model = LinearRegression()
Model.fit(X,Y)

#predict values
test_data = [[1800], [3000]]
predictions = Model.predict(test_data)

print("Predicted Prices:")
print(predictions)
```

```python
import pandas as pd

# Dataset
data = {
    'Category': ['A', 'B', 'C', 'A', 'D', 'B', 'A', 'E', 'B', 'C', 'C', 'A'],
    'Value': [10, 15, 20, 30, 25, 18, 22, 40, 35, 50, 45, 15]
}

# Create DataFrame
df = pd.DataFrame(data)

# Count occurrences of each category
category_counts = df['Category'].value_counts()

# Get top 3 categories
top_categories = category_counts.head(3)

print("Category counts:")
print(category_counts)
print("\nTop 3 most frequent categories:")
print(top_categories)
```

```python
import pandas as pd

# Dataset
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, None, 35, None]}
# Create DataFrame
df = pd.DataFrame(data)
print("Data with NA values")
print(df)
df['Age'].fillna(df['Age'].mean(), inplace=True)
print("Data after imputation")
print(df)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


#Dataset:
data = {'Experience': [1, 2, 3, 4, 5], 
        'Salary': [30000, 35000, 40000, 45000, 50000]}

df = pd.DataFrame(data)
df

#splitting features & target
X = df[['Experience']]
Y = df['Salary']

#Training linear regression 
Model = LinearRegression()
Model.fit(X,Y)

#predict values
test_data = [[6]]
predictions = Model.predict(test_data)

print("Expected Salary:")
print(predictions)
```
