```python
#8.
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer


# Creating a dataset with missing values
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

# Create an imputer to handle missing values
# strategy='mean' replaces NaN with the mean of the column
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on X_train and transform X_train
X_train_imputed = imputer.fit_transform(X_train)


model = LinearRegression()
model.fit(X_train_imputed, y_train)  
print(model.predict([[2,3]]))
```
